```python
import os
import random
import string
os.chdir("./Images")
```

```python
images=os.listdir(os.getcwd())
letters=[char for char in string.ascii_lowercase]
def generate_name():
    name=""
    num=random.randint(1,len(images))
    for i in range(1,5):
        new_name=random.choice(letters)
        name+=new_name
    return f"{name}_{num}"

for image in images:
    old_name=image
    new_name=generate_name()
    if old_name.split(".")[-1].lower().endswith(('jpeg','png','jpg')):
        os.rename(old_name,f"{new_name}.jpeg")
#         print(f"renamed {old_name} to {new_name}")
else:
    print("All Done!")
    
    
images
```
